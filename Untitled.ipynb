```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import tensorflow as tf

from glob import glob
from tqdm import tqdm

import matplotlib.pyplot as plt
from PIL import Image
```

```python
open_eyes_dir = 'C:/Users\Vibhuti Sharma/OneDrive - DIT University/Desktop/drowsiness project/train/Open_Eyes'
closed_eyes_dir = 'C:/Users\Vibhuti Sharma/OneDrive - DIT University/Desktop/drowsiness project/train/Closed_Eyes'
```

```python
X = []
Y = []

for i in tqdm(os.listdir(open_eyes_dir)):
        if i.endswith('.jpg') or i.endswith('.png'):  # Assuming images are in JPEG or PNG format
            temp = np.array(Image.open(os.path.join(open_eyes_dir, i)).resize((64, 64)))
            X.append(temp)
            Y.append(1)
    
for i in tqdm(os.listdir(closed_eyes_dir)):
        if i.endswith('.jpg') or i.endswith('.png'):
            temp = np.array(Image.open(os.path.join(closed_eyes_dir, i)).resize((64, 64)))
            X.append(temp)
            Y.append(0)  # Label for closed eyes images
    
```

```python
X = np.array(X)
X = X/255.0
Y = np.array(Y)
```

```python
X.shape
```

```python
X = np.expand_dims(X,-1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Conv2D, BatchNormalization, MaxPooling2D,Dropout, Flatten
```

```python
model = tf.keras.models.Sequential([
      Input(shape=(64, 64, 1)),

      Conv2D(filters = 32, kernel_size = 5, strides = 1, activation = 'relu'),
      Conv2D(filters = 32, kernel_size = 5, strides = 1, activation = 'relu', use_bias=False),
      BatchNormalization(),
      MaxPooling2D(strides = 2),
      Dropout(0.3),

      Conv2D(filters = 64, kernel_size = 3, strides = 1, activation = 'relu'),
      Conv2D(filters = 64, kernel_size = 3, strides = 1, activation = 'relu', use_bias=False),
      BatchNormalization(),
      MaxPooling2D(strides = 2),
      Dropout(0.3),

      Flatten(),
      Dense(units  = 256, activation = 'relu', use_bias=False),
      BatchNormalization(),

      Dense(units = 128, use_bias=False, activation = 'relu'),

      Dense(units = 84, use_bias=False, activation = 'relu'),
      BatchNormalization(),
      Dropout(0.3),

      Dense(units = 1, activation = 'sigmoid')
  ])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='bestModel.h5',
    save_weights_only=False,
    monitor='val_loss',
    mode='min',
    save_best_only=True,
    verbose =1)
model.fit(x_train, y_train, validation_split=0.2, epochs=30, batch_size=32, callbacks=callback)
```

```python
model.evaluate(x_test, y_test)
```

```python
from keras.models import load_model
best_model = load_model('C:/Users/Vibhuti Sharma/OneDrive - DIT University/Desktop/drowsiness project/bestModel.h5')
best_model.evaluate(x_test, y_test)
```

```python
for i in x_test[0:5]:
    result = best_model.predict(np.expand_dims(i,0))
    plt.imshow(i)
    plt.show()
    
    if result > 0.5:
        print('Open')
    else:
        print("Closed")
    
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

plt.figure(figsize=(15, 5))

preds = best_model.predict(x_test)
preds = (preds >= 0.5).astype(np.int32)
cm = confusion_matrix(y_test, preds)
df_cm = pd.DataFrame(cm, index=['closed', 'Open'], columns=['Closed', 'Open'])
plt.subplot(121)
plt.title("Confusion matrix\n")
sns.heatmap(df_cm, annot=True, fmt="d", cmap="YlGnBu")
plt.ylabel("Predicted")
plt.xlabel("Actual")
```
